```python
#14 Is there any visible relationship between box office earnings and movie ratings?
data = df.groupby('Rating')['Revenue (Millions)'].agg(['mean'])
data_dict = data.to_dict()['mean']
x = data_dict.keys()
y = data_dict.values()

figure = plt.figure(figsize = (15,5))
plt.plot(x,y)
plt.xlabel('Ratings')
plt.ylabel('Revenue (Millions) (mean)')
plt.show()
```

```python
#20 Plot the average movie ratings over the years to identify trends.
data = df.groupby('Year')['Rating'].agg(['median'])
data_dict = data.to_dict()['median']

x = data_dict.keys()
y = data_dict.values()

figure = plt.figure(figsize = (15,5))
plt.plot(x,y)
plt.xlabel('year')
plt.ylabel('rating(mean)')
plt.title('average movie ratings over the years')
plt.show()
```

```python
# 23. Analyze how movie runtimes have changed over the years. Are movies getting longer or shorter?
data = df.groupby('Year')['Runtime (Minutes)'].agg(['median'])
data_dict = data.to_dict()['median']

x = data_dict.keys()
y = data_dict.values()

figure = plt.figure(figsize=(15,5))
plt.plot(x,y)
plt.xlabel('years')
plt.ylabel('runtime (in minutes) (mean)')
plt.title('how movie runtimes have changed over the years')
plt.show()
```

```python
# Q.19 19. Create a word cloud visualization using movie titles or keywords from descriptions

from wordcloud import WordCloud

titles = (' ').join(df['Title'])

wordcloud = WordCloud(height = 500, width=500, background_color = 'black').generate(titles)

figure = plt.figure(figsize=(15,5))
plt.imshow(wordcloud)
plt.axis('off')
plt.show()

```
